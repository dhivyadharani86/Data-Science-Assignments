Descriptive Analysis

Sales Trend Over Time

```python
sales_trend = df.groupby('Day')['Total Sales Value'].sum()
# Plot the trend
plt.figure(figsize=(10,6))
plt.plot(sales_trend.index, sales_trend.values, label='Sales Over Time')
plt.xlabel('Day')
plt.ylabel('Total Sales Value')
plt.title('Sales Trend Over Time')
plt.legend()
plt.show()
```

Total sales per BU

```python
# Total sales per BU
sales_by_bu = df.groupby('BU')['Total Sales Value'].sum().sort_values()
# Plot total sales by product
plt.figure(figsize=(10,6))
sales_by_bu.plot(kind='barh')
plt.xlabel('Total Sales Value')
plt.ylabel('BU')
plt.title('Sales by BU')
plt.show()
```

Total sales per Brand

```python
# Total sales per Brand
sales_by_brand = df.groupby('Brand')['Total Sales Value'].sum().sort_values()
# Plot total sales by product
plt.figure(figsize=(10,6))
sales_by_brand.plot(kind='barh')
plt.xlabel('Total Sales Value')
plt.ylabel('Brand')
plt.title('Sales by Brand')
plt.show()
```

```python
# Select only the numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Set up the plot size and layout
plt.figure(figsize=(15, 12))

# Loop through each numerical column and plot a histogram
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns) // 3 + 1, 3, i)
    sns.histplot(df[col], kde=True, bins=30)  # KDE adds a smooth curve over the histogram
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Calculate skewness and kurtosis for each numerical column

```python
# Calculate skewness and kurtosis for each numerical column
for col in numerical_columns:
    skewness = df[col].skew()
    kurtosis = df[col].kurt()
    print(f'{col}: Skewness = {skewness:.2f}, Kurtosis = {kurtosis:.2f}')
```

Plot box plot for each numerical column

```python
# Plot box plot for each numerical column
plt.figure(figsize=(15, 8))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns) // 3 + 1, 3, i)
    sns.boxplot(y=df[col], color='lightblue')
    plt.title(f'Box plot of {col}')

plt.tight_layout()
plt.show()
```

Identify Categorical Columns

```python
# Select only the categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
```

```python
# Set up the plot size and layout
plt.figure(figsize=(15, 10))

# Loop through each categorical column and plot a bar chart
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(len(categorical_columns) // 2 + 1, 2, i)
    sns.countplot(y=col, data=df, palette='viridis')  # y is used for horizontal bar chart
    plt.title(f'Frequency of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

Standardize the Numerical Columns

```python
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()

# Apply the scaler on numerical columns and return a standardized version
df_standardized = df.copy()  # To keep the original dataset
df_standardized[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Display the first few rows of the standardized dataset
print(df_standardized.head())
```

Visualize Standardization Effect (Before and After)

```python
# Plot histograms of original numerical data
plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns) // 3 + 1, 3, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Original {col}')
    
plt.tight_layout()
plt.show()
```

Histogram of Standardized Data:

```python
# Plot histograms of standardized numerical data
plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns) // 3 + 1, 3, i)
    sns.histplot(df_standardized[col], kde=True, bins=30)
    plt.title(f'Standardized {col}')
    
plt.tight_layout()
plt.show()
```

```python
for col in numerical_columns:
    print(f'{col} - Mean: {df_standardized[col].mean():.2f}, Std Dev: {df_standardized[col].std():.2f}')
```

```python
# Apply one-hot encoding to categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=False)  # drop_first=False keeps all dummy columns
# Display the first few rows of the transformed dataset
print(df_encoded.head())
```
